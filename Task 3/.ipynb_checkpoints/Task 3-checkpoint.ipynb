## Tensorflow

```python
import tensorflow as tf
from tensorflow.keras import datasets
import datetime as dt
from tensorflow import contrib

tf.enable_eager_execution()
```

```python
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
shuffle_index = np.random.choice(len(y_train), len(y_train))

train_dataset = tf.data.Dataset.from_tensor_slices(
    (
        x_train,
        x_train[shuffle_index],
        y_train,
        y_train[shuffle_index]
    )
).batch(32).shuffle(10000)

train_dataset = train_dataset.map(
    lambda x_l, x_r, y_l, y_r: (
        tf.reshape(tf.div(tf.cast(x_l, tf.float32), 255.0), (-1, 28, 28, 1)),
        tf.reshape(tf.div(tf.cast(x_r, tf.float32), 255.0), (-1, 28, 28, 1)),
#         tf.reshape(tf.one_hot(y_l, 10), (-1, 10))
        tf.one_hot(tf.dtypes.cast(tf.math.equal(y_l, y_r), tf.int64), 2)
    )
)

train_dataset = train_dataset.map(
    lambda x_l, x_r, y: (
        tf.image.random_flip_left_right(x_l),
        tf.image.random_flip_left_right(x_r),
        y
    )
)

train_dataset = train_dataset.repeat()
```

```python
test_shuffle_index = np.random.choice(len(y_test), len(y_test))

test_dataset = tf.data.Dataset.from_tensor_slices(
    (
        x_test,
        x_test[test_shuffle_index],
        y_test,
        y_test[test_shuffle_index]
    )
).batch(100).shuffle(10000)

test_dataset = test_dataset.map(
    lambda x_l, x_r, y_l, y_r: (
        tf.reshape(tf.div(tf.cast(x_l, tf.float32), 255.0), (-1, 28, 28, 1)),
        tf.reshape(tf.div(tf.cast(x_r, tf.float32), 255.0), (-1, 28, 28, 1)),
#         tf.reshape(tf.one_hot(y_l, 10), (-1, 10))
        tf.one_hot(tf.dtypes.cast(tf.math.equal(y_l, y_r), tf.int64), 2)
    )
)

test_dataset = test_dataset.map(
    lambda x_l, x_r, y: (
        tf.image.random_flip_left_right(x_l),
        tf.image.random_flip_left_right(x_r),
        y
    )
)

test_dataset = test_dataset.repeat()
```

```python
class SiameseModel(tf.keras.models.Model):
    def __init__(self, input_shape=(28, 28)):
        super(SiameseModel, self).__init__(name='siamese')
        # Left
        self.conv_left_1 = tf.keras.layers.Conv2D(64, 5,
                                         padding='same',
                                         activation=tf.nn.relu,
                                         kernel_initializer=tf.initializers.variance_scaling,
                                         kernel_regularizer=tf.keras.regularizers.l2(l=0.001),
                                         input_shape=(input_shape[0], input_shape[1], 1)
                                        )
        self.max_pool2d = tf.keras.layers.MaxPooling2D((3, 3), (2, 2), padding='same')
        self.max_norm = tf.keras.layers.BatchNormalization()
        self.conv_left_2 = tf.keras.layers.Conv2D(64, 5,
                                         padding='same',
                                         activation=tf.nn.relu,
                                         kernel_initializer=tf.initializers.variance_scaling,
                                         kernel_regularizer=tf.keras.regularizers.l2(l=0.001))
        self.flatten = tf.keras.layers.Flatten()
        self.left_neck = tf.keras.layers.Dense(128, activation=tf.nn.relu,
                                      kernel_initializer=tf.initializers.variance_scaling,
                                      kernel_regularizer=tf.keras.regularizers.l2(l=0.001))
        
        # Right
        self.conv_right_1 = tf.keras.layers.Conv2D(64, 5,
                                         padding='same',
                                         activation=tf.nn.relu,
                                         kernel_initializer=tf.initializers.variance_scaling,
                                         kernel_regularizer=tf.keras.regularizers.l2(l=0.001),
                                         input_shape=(input_shape[0], input_shape[1], 1)
                                        )
        self.conv_right_2 = tf.keras.layers.Conv2D(64, 5,
                                         padding='same',
                                         activation=tf.nn.relu,
                                         kernel_initializer=tf.initializers.variance_scaling,
                                         kernel_regularizer=tf.keras.regularizers.l2(l=0.001))
        self.right_neck = tf.keras.layers.Dense(128, activation=tf.nn.relu,
                                      kernel_initializer=tf.initializers.variance_scaling,
                                      kernel_regularizer=tf.keras.regularizers.l2(l=0.001))
        
        # Join
        self.neck = tf.keras.layers.Dense(128, activation=tf.nn.relu,
                                      kernel_initializer=tf.initializers.variance_scaling,
                                      kernel_regularizer=tf.keras.regularizers.l2(l=0.001))
        self.dropout = tf.keras.layers.Dropout(0.5)
        self.logits = tf.keras.layers.Dense(2)
        self.softmax = tf.keras.layers.Softmax()

    def call(self, left, right):
        # Left
        left = self.max_pool2d(self.conv_left_1(left))
        left = self.max_norm(left)
        left = self.max_pool2d(self.conv_left_2(left))
        left = self.max_norm(left)
        left = self.flatten(left)
        left = self.left_neck(left)
        
        # Left
        right = self.max_pool2d(self.conv_right_1(right))
        right = self.max_norm(right)
        right = self.max_pool2d(self.conv_right_2(right))
        right = self.max_norm(right)
        right = self.flatten(right)
        right = self.right_neck(right)
        
        x = tf.keras.layers.concatenate([left, right])
        x = self.dropout(self.neck(x))
        x = self.logits(x)
        return self.softmax(x)
```

```python
tfe = contrib.eager

model = SiameseModel()
model.compile(optimizer=tf.train.AdamOptimizer(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# keep results for plotting
train_loss_results = []
train_accuracy_results = []

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
global_step = tf.Variable(0)

num_epochs = 201

for epoch in range(num_epochs):
    epoch_loss_avg = tfe.metrics.Mean()
    epoch_accuracy = tfe.metrics.Accuracy()
        
    x_l, x_r, y = next(iter(train_dataset))
    with tf.GradientTape() as tape:
        logits = model(x_l, x_r)
        loss = tf.losses.softmax_cross_entropy(
            y,
            logits,
        )
        
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables), global_step)

    # Track progress
    epoch_loss_avg(loss)  # add current batch loss
    # compare predicted label to actual label
    acc_pred = tf.argmax(model(x_l, x_r), axis=1, output_type=tf.int32)
    acc_label = tf.argmax(y, axis=1, output_type=tf.int32)
    epoch_accuracy(acc_pred, acc_label)

    # end epoch
    train_loss_results.append(epoch_loss_avg.result())
    train_accuracy_results.append(epoch_accuracy.result())
  
    if epoch % 50 == 0:
        print("Epoch {:03d}: Loss: {:.3f}, Accuracy: {:.3%}".format(epoch,
                                                                epoch_loss_avg.result(),
                                                                epoch_accuracy.result()))
```

```python
image_path = "seen-dataset/TrainingSet/"
batch_size = 32
writer_samples = list(set(list(df_seen.right.unique()) + list(df_seen.left.unique())))
writer_sample_index = 5
loaded_image = cv2.imread(os.path.join(image_path, writer_samples[writer_sample_index]), 0)
plt.imshow(loaded_image)
```

```python
def gen():
    left_indexes = np.random.randint(0, len(writer_samples), batch_size)
    right_indexes = np.random.randint(0, len(writer_samples), batch_size)
    x_l, x_r, y = [], [], []
    for left_index, right_index in zip(left_indexes, right_indexes):
        left_image = cv2.imread(os.path.join(image_path, writer_samples[left_index]), 0)
        right_image = cv2.imread(os.path.join(image_path, writer_samples[right_index]), 0)
        
        x_l.append(left_image)
        x_r.append(right_image)
        y.append(writer_samples[left_index][:4] == writer_samples[right_index][:4])
    
    yield (x_l, x_r, y)

ds = tf.data.Dataset.from_generator(gen, (tf.float64, tf.float64, tf.int64))
```

```python
train_dataset = ds.map(
    lambda x_l, x_r, y: (
        tf.reshape(tf.div(tf.cast(x_l, tf.float32), 255.0), (-1, 64, 64, 1)),
        tf.reshape(tf.div(tf.cast(x_r, tf.float32), 255.0), (-1, 64, 64, 1)),
        tf.one_hot(y, 2)
    )
)

train_dataset = train_dataset.map(
    lambda x_l, x_r, y: (
        tf.image.random_flip_left_right(x_l),
        tf.image.random_flip_left_right(x_r),
        y
    )
)

train_dataset = train_dataset.repeat()
```
