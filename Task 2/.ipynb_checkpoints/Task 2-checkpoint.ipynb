# Testing Hypothesis

• $h_0$: Evidence (with characteristics) $e$ was generated by known (individual with characteristics) $k$

• $h_1$: Evidence $e$ was not generated by $k$

## Likelihood Ratio: $LR(k,e) = \frac{p(k,e | h_0)}{p( k,e | h_1)}$

```python
df_raw['writer_id'] = df_raw.ImageId.str.replace('[a-z]', '')
```

```python
df = df_raw.copy()
df['key'] = 1
df = pd.merge(df, df, on='key')
del df['key']
df['h'] = (df['writer_id_x'] != df['writer_id_y']).astype(int)
```

```python
node_index = df.columns[df.columns.str.contains('f._|h')]

def get_blank_model():
    model = BayesianModel()
    model.add_nodes_from(node_index)
    return model

def tuples_to_edges(tuples):
    edges = []
    for z in ['x', 'y']:
        edges += [
            (f'f{u}_{z}', f'f{v}_{z}') if v > 0 else (f'f{u}_{z}', f'h')
            for (u, v) in tuples
        ]
    return edges

def generate_model(tuples):
    model = get_blank_model()
    edges = tuples_to_edges(tuples)
    model.add_edges_from(edges)
    model.fit(df)
    return model

# Get Subset of Data
df_0 = df[df.h == 0]
df_1 = df.iloc[np.random.choice(df[df.h == 1].shape[0], size=len(df_0), replace=False)].sort_index()
df_sample = pd.concat([df_0, df_1]).copy()
```

```python
model = generate_model([(1, 2), (3, 4), (2, 5), (4, 5), (6, 7), (8, 9), (7, 9), (5, 0), (9, 0)])
```

## Sampling

```python
from pgmpy.sampling import BayesianModelSampling
from pgmpy.utils.mathext import sample_discrete
import itertools
from pgmpy.sampling import _return_samples
from tqdm import tqdm
```

```python
sampler = BayesianModelSampling(k_model)
```

```python
%%time
samples = sampler.forward_sample(size=1)
```

```python
size = 1
topological_order = list(nx.topological_sort(k_model))
types = [(var_name, 'int') for var_name in topological_order]
sampled = np.zeros(size, dtype=types).view(np.recarray)

cardinality = {}
for node in k_model.nodes():
    cpd = k_model.get_cpds(node)
    cardinality[node] = cpd.variable_card
    
def pre_compute_reduce(variable):
    variable_cpd = k_model.get_cpds(variable)
    variable_evid = variable_cpd.variables[:0:-1]
    cached_values = {}

    count = 0
    for state_combination in itertools.product(*[range(cardinality[var]) for var in variable_evid]):
        count += 1
        
    for state_combination in tqdm(itertools.product(*[range(cardinality[var]) for var in variable_evid]), total=count):
        states = list(zip(variable_evid, state_combination))
        cached_values[state_combination] = variable_cpd.reduce(states, inplace=False).values

    return cached_values
```

```python
for node in topological_order:
    print(node)
    cpd = k_model.get_cpds(node)
    states = range(cardinality[node])
    evidence = cpd.variables[:0:-1]
    if evidence:
        cached_values = pre_compute_reduce(variable=node)
        evidence = np.vstack([sampled[i] for i in evidence])
        weights = list(map(lambda t: cached_values[tuple(t)], evidence.T))
    else:
        weights = cpd.values
    sampled[node] = sample_discrete(states, weights, size)
```

```python
pre_compute_reduce(variable=node)
```

```python
_return_samples('dataframe', sampled)
```
